### c) Broj uknjiženih/neuknjiženih nekretnina

```python
uknjizene_kuce_count = len(df[(df['uknjizenost'] == 'Da') & (df['tip_nekretnine'] == 'Kuća')])
neuknjizene_kuce_count = len(df[(df['uknjizenost'] == 'Ne') & (df['tip_nekretnine'] == 'Kuća')])

uknjizeni_stanovi_count = len(df[(df['uknjizenost'] == 'Da') & (df['tip_nekretnine'] == 'Stan')])
neuknjizeni_stanovi_count = len(df[(df['uknjizenost'] == 'Ne') & (df['tip_nekretnine'] == 'Stan')])
```

```python
print('Broj uknjizenih kuca: ', uknjizene_kuce_count)
print('Broj neuknjizenih kuca: ', neuknjizene_kuce_count)
print('')
print('Broj uknjizenih stanova: ', uknjizeni_stanovi_count)
print('Broj neuknjizenih stanova: ', neuknjizeni_stanovi_count)
```
